```python
import pandas as pd

data = pd.read_excel("ideal.xlsx")
data.tail()
```

```python
r = ""
for index, row in data.iterrows():
    umur = row["Umur"]
    berat = row["Berat Ideal (Kg)"]
    tinggi = row["Tinggi Ideal ()"]
    lingkar = row["Lingkar Kepala ()"]

    berat_ = berat.split("-")
    r += "\n"
    r += f'aData = new ArrayList<>();aData.add({berat_[0].strip()});aData.add({berat_[1].strip()});ref.put("GENDER_MALE BERAT {umur}", aData);'
    tinggi_ = tinggi.split("-")
    r += "\n"
    r += f'aData = new ArrayList<>();aData.add({tinggi_[0].strip()});aData.add({tinggi_[1].strip()});ref.put("GENDER_MALE TINGGI {umur}", aData);'
    lingkar_ = lingkar.split("-")
    r += "\n"
    r += f'aData = new ArrayList<>();aData.add({lingkar_[0].strip()});aData.add({lingkar_[1].strip()});ref.put("GENDER_MALE KEPALA {umur}", aData);'

print(r)
```
